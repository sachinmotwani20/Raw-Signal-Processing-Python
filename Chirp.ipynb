#Chirp
A chirp is a signal with changing frequency in time.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
start_freq = 1
end_freq = 100
samples = 44100
time_basis = np.linspace(0, np.pi * 2, samples)
chirp = np.linspace(start_freq, end_freq, samples) #Chirp in terms of frequency
signal = np.cos(time_basis * chirp) #Chirp in terms of time
```

```
plt.plot(time_basis, signal)
```
