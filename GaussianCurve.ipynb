#Gaussian Curve 
In one dimension, the Gaussian curve is the probability density function of the **normal distribution** 
![alt text](https://drive.google.com/uc?id=1AnNDaPSLqH_npTWARrEYYTWlTSr6OIxb)



```
#Mixed Signals.Learn
import numpy as np
import matplotlib.pyplot as plt
```

```
#Method 1: Direct Formula

srate = 1000 #Sampling Rate
time = np.arange(-2.0,2.0, 1/srate) #Total 4000 points

#Signal 
gauss = np.exp((-time**2)/0.1) 

#Plotting
plt.plot(time, gauss)
plt.title('1D Gaussian Distribution Curve')
plt.show()
```

```
#Method 2: Using scipy
import scipy.stats


mean = 0
std_dvn = 1


time = np.arange(-2.0, 2.0, 0.001) #Another way of declaring time

#Signal
gauss = scipy.stats.norm(mean, std_dvn)

#plot
plt.plot(time, gauss.pdf(time))  #Notice PDF
plt.title('1D Gaussian Distribution Curve')
plt.show()
```

In two dimensions, the circular Gaussian function is the distribution function for uncorrelated variates X and Y having a bivariate normal distribution and equal standard deviation  σ= σx= σy 

![alt text](https://drive.google.com/uc?id=172SylSMF9vkPs94ZMLkRQYqLwk4C5GZg)

```
#Plotting 2D Gaussian with equal standard deviations
size = 1000
sigma_x = 6.
sigma_y = 6.

x = np.linspace(-10, 10, size)
y = np.linspace(-10, 10, size)

x, y = np.meshgrid(x, y) #used to create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing
z = (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x**2/(2*sigma_x**2) + y**2/(2*sigma_y**2))))

plt.contourf(x, y, z, cmap='Blues')
plt.colorbar()
plt.title('2D Gaussian with equal standard deviations')
plt.show()
```

```
#Plotting 2D Gaussian of 2 different standard deviations (sigma)
size = 100
sigma_x = 6.
sigma_y = 2.

x = np.linspace(-10, 10, size)
y = np.linspace(-10, 10, size)

x, y = np.meshgrid(x, y)
z = (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x**2/(2*sigma_x**2)
     + y**2/(2*sigma_y**2))))

plt.contourf(x, y, z, cmap='Greens')
plt.colorbar()
plt.title('2D Gaussian of 2 different standard deviations')
plt.show()
```
