Here f1 represents f(t) & g1 represetns g(t.

For Simpicity, x=1.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
t = np.arange(-1.0, 5.0, 0.01)
phas = 0*np.pi/4
g1 = 5*np.sin(2*np.pi*t + phas) #Sine Function

plt.plot(t,g1)
plt.grid(True)
plt.show()
```

```
g1.shape
```

```
t.shape
```

```
f1 = np.array([np.exp(-i) * j if i > 0 else 0 for i,j in zip(t,g1) ])
```

```
f1.shape
```

Done the above shape thing to undertand/debug the ploting problem [Now Resolved; kept for reference]

```
plt.plot(t,f1)
plt.grid(True)
plt.show()
```

#Fourier Transform of f1

```
freq = np.fft.fftfreq(t.shape[-1])
f1_fft = np.fft.fft(f1)
plt.plot(freq, f1_fft.real)
plt.xlim([-0.05, 0.05])
plt.show()
plt.plot( freq, f1_fft.imag)
plt.xlim([-0.05, 0.05])
plt.show()
```

```
plt.plot(freq, f1_fft)
plt.xlim([-0.05, 0.05])
plt.show()
```

#Try sawtooth instead of Sine Wave

```
from scipy import signal

t = np.linspace(0, 1, 500)
Sawtooth = 5*signal.sawtooth(2 * np.pi * 5 * t) #Sawtooth Function
plt.plot(t, Sawtooth, 'r')
plt.grid(True)
plt.show()
```
