#Morlet Wavelet (Gabor Wavelet) 
It is a wavelet composed of a complex exponential (carrier) multiplied by a Gaussian window (envelope). This wavelet is closely related to human perception, both hearing and vision.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
#phase
theta = 3*np.pi/4

#simulation parameters 
srate = 1000 #Sampling Rate
time = np.arange(-1.0, 1.0,1/srate )

#Signals 
sine = 5 *np.sin(2*np.pi*6*time + theta)   #Carrier
gauss = np.exp((-time**2)/0.1) #envelop

morlet = np.multiply(sine, gauss)  #Morlet Wavelet : It's a sine Wave tapered by a Gaussian

plt.plot(time, sine)
plt.title('Sine Wave')
plt.show()

plt.plot(time, gauss)
plt.title('Gaussian Curve')
plt.show()

plt.plot(time, morlet)
plt.title('Morlet Wavelet')
plt.show()
```
