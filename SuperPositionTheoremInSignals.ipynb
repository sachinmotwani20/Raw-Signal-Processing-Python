```
#SuperPosition Theorem
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-1.0,5.0, 0.02)
```

```
#Sine Wave
sin_wave = 5*np.sin(2*np.pi*t)
plt.plot(t, sin_wave)
plt.title ('sin(x)')
plt.grid(True)
plt.show()
```

```
#Cos Wave
cos_wave = 3*np.cos(2*np.pi*t)
plt.plot(t, cos_wave,'r')
plt.title ('cos(x)')
plt.grid(True)
plt.show()
```

```
#Additivity -> Scalling
mixed_wave = 6*(sin_wave + cos_wave)
plt.plot(t, mixed_wave, 'g')
plt.title('6[sin(x)+ cos(x)]')
plt.grid(True)
plt.show()
```

```
#Scalling -> Additivity
mixed_wave1= (6*sin_wave + 6*cos_wave)
plt.plot(t, mixed_wave, 'y')
plt.title('[6(sin(x)) + 6(cos(x)]')
plt.grid(True)
plt.show()
```

```
#Hence Superposition Theorem has been Proved.
```
