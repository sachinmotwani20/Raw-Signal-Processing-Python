#How to Zoom into a Graph
This notebook desribes, in detail, how one can zoom into graphs in Google Colab.It uses a simple API for exporting the matplotlib graphics to HTML code which can be used within the browser, within standard web pages, blogs, or tools such as the google colab & IPython notebook.

mpld3 is used for creating interactive data visualizations for the web. 
It is used creating interactive graphs for libraries like matplotlib & D3js.

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import mpld3
from mpld3 import plugins

# Define some CSS to control our custom labels (Optional Enhancement)
css = """
table
{
  border-collapse: collapse;
}
th
{
  color: #ffffff;
  background-color: #000000;
}
td
{
  background-color: #cccccc;
}
table, th, td
{
  font-family:Arial, Helvetica, sans-serif;
  border: 1px solid black;
  text-align: right;
}
"""
#Why this?
#The by-default table is not pleasing to eyes.


#Creating grid for the plot
fig, ax = plt.subplots()
ax.grid(True, alpha=0.3)

#Generating Random Data
N = 50
df = pd.DataFrame(index=range(N))
df['x'] = np.random.randn(N)
df['y'] = np.random.randn(N)
df['z'] = np.random.randn(N)

#For Locating datapoint
labels = []
for i in range(N):
    label = df.iloc[[i], :].T
    label.columns = ['Row {0}'.format(i)]
    # .to_html() is unicode; so make leading 'u' go away with str()
    labels.append(str(label.to_html()))

#Plotting Points 
points = ax.plot(df.x, df.y, 'o', color='b', mec='k', ms=15, mew=1, alpha=.6)

#Title & Axes labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title', size=20)

tooltip = plugins.PointHTMLTooltip(points[0], labels, voffset=10, hoffset=10, css = css)
plugins.connect(fig, tooltip)

mpld3.display()
```
